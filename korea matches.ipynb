```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('WorldCupMatches.csv')
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
df1 = df.loc[(df['Home Team Name'] == 'Korea Republic') | (df['Away Team Name'] == 'Korea Republic')]
```

```python
df2=df1[['Year','Datetime','City','Stage','Home Team Name','Away Team Name','Home Team Goals','Away Team Goals']]
```

```python
df2['Korea Score'] = 0
```

```python
df2.head(1)
```

```python
df2['Korea Score'] = np.where(df2['Home Team Name'] == 'Korea Republic', df2['Home Team Goals']\
                             ,np.where(df2['Away Team Name'] == 'Korea Republic',df2['Away Team Goals'],0))
```

```python
df2['VS Score'] = np.where(df2['Home Team Name'] == 'Korea Republic', df2['Away Team Goals']\
                          ,np.where(df2['Away Team Name'] == 'Korea Republic',df2['Home Team Goals'],0))
```

```python
df2['Result'] = df2['Korea Score'] - df2['VS Score']
```

```python
df2['Win'] = np.where(df2['Result'] > 0 ,1,0)
df2['Loss'] = np.where(df2['Result'] < 0 ,1,0)
df2['Equal'] = np.where(df2['Result'] == 0 ,1,0)
```

```python
df2.head(1)
```

```python
df2.groupby('Year').agg({'Result':'sum'})
```

```python
df2.groupby('Year').agg({'Win':'sum'})
```

```python
df2['Opponent'] = np.where(df2['Home Team Name'] == 'Korea Republic' , df2['Away Team Name'], df2['Home Team Name'])
```

```python
df2['>10'] = np.where(df2['Opponent'] == 'Sweden',1\
                     ,np.where(df2['Opponent']=='Germany',1\
                     ,np.where(df2['Opponent']=='Brazil',1\
                     ,np.where(df2['Opponent']=='Argentina',1\
                     ,np.where(df2['Opponent']=='England',1\
                     ,np.where(df2['Opponent']=='France',1\
                     ,np.where(df2['Opponent']=='Spain',1\
                     ,np.where(df2['Opponent']=='Italy',1,0))))))))
```
